```python
def largo(texto):
    resultado = 0
    for _ in texto:
        resultado += 1
    return resultado
b = largo("chanchito feliz")
print(b)
```

```python
def suma(a, b):
    resultado = a + b
    return resultado
c = suma(1, 2)
suma(c, 2)
```
