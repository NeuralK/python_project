```python
# Una lista de los cubos del 1 al 10, mediante comprension de listas.

numeros = [patata**3 for patata in range(1, 11)]
print(f"Este es el resultado: {numeros}")
```

```python
# Iteracion simple con listas.

pizzas = ["pizza italiana", "pizza de muzarella", "pizza con afaina", "pizza con panceta"]

for pizza in pizzas:
    print(f"Esta {pizza.title()} es de las mejores del mundo.")
print("Soy un desquiciado que ama las pizzas XD.")
```
