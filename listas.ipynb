```python
usuarios = [[1, "felipe"], [4, "chanchon"], [2, "weon"], [0, "error"]]

def ordena(element):
    return element[0]
usuarios.sort(key=ordena)
print(usuarios)
usuarios.sort(key=ordena, reverse=True)
print(usuarios)
```
