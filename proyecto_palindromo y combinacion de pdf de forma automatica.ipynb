```python
def palabra(texto):
    texto = texto.lower().replace(" ", "")
    invertir = ""

    
    for chinche in texto:
        invertir = chinche + invertir

    if texto == invertir:
        return True
    else:
        return False

print(palabra("RAdar"))
```

```python
import PyPDF2
import os
import re

# --- CONFIGURACIÓN ---
# Cambiá esto por la carpeta donde están tus archivos (ej: "C:/Escaneos/Mama")
carpeta_insumos = r"C:\Users\Usuario\Desktop\Carta ingeniero" 
nombre_final = "Documento_Combinado_Final.pdf"

def obtener_numero(nombre_archivo):
    # Busca cualquier número en el nombre. '0001' se convierte en el entero 1.
    match = re.search(r'\d+', nombre_archivo)
    return int(match.group()) if match else 0

def ejecutar_fusion():
    merger = PyPDF2.PdfMerger()
    
    # 1. Filtramos solo los archivos que terminan en .pdf
    lista_pdfs = [f for f in os.listdir(carpeta_insumos) if f.endswith('.pdf')]
    
    # 2. Ordenamos la lista usando nuestra función de números
    # Esto asegura que '0002' venga después de '0001'
    lista_pdfs.sort(key=obtener_numero)
    
    print(f"Detectados {len(lista_pdfs)} archivos. Iniciando fusión...")

    # 3. Pegamos uno tras otro
    for archivo in lista_pdfs:
        ruta_completa = os.path.join(carpeta_insumos, archivo)
        print(f"Añadiendo: {archivo}")
        merger.append(ruta_completa)
    
    # 4. Guardamos el resultado
    merger.write(nombre_final)
    merger.close()
    print(f"\n✅ ¡PROCESO COMPLETADO! Archivo generado: {nombre_final}")

if __name__ == "__main__":
    ejecutar_fusion()
```

```python
def palindromo(texto):
    limpio = texto.lower().replace(" ", "")
    return limpio == texto[::-1]

print(palindromo("salame"))
print(palindromo("radar"))
print()
```
