```python
class Usuario:

    def __init__(self, nombre, apellido, age):
        
        self.nombre = nombre

        self.apellido = apellido

        self.age = int(age)

    def describir_usuario(self):

        print(f"\nAsí que usted se llama {self.nombre}.")
        print(f"\nMmmh te apellidas {self.apellido} raro apellido.") 
        print(f"\nAsí que tienes {self.age} es curioso.")
    

    def saludar(self):

        self.saludo = "\nBienvenido"

        print(self.saludo.title())


    
banana = Usuario("Kevin", "Moreno", 17)

banana.describir_usuario()

banana.saludar()

anana = Usuario("Luis", "Boreno", 37)

anana.describir_usuario()

anana.saludar()
```

```python
class Restaurante:

    def __init__(self, nombre_restaurante, tipo_cocina):
        self.nombre_restaurante = nombre_restaurante
        self.tipo_cocina = tipo_cocina

    def descripcion_restaurante(self):

        print(f"Así que el restaurante se llama {self.nombre_restaurante}.")
        print(f"\nAsí que {self.tipo_cocina} es el tipo de cocina de usan.")
    
    def abrir_restaurante(self):
        self.open = "\nRestaurante abierto"
        
        print(self.open.title())

banana = Restaurante("Buiton", "Oriental")

banana.descripcion_restaurante()

banana.abrir_restaurante()

anana = Restaurante("Salame", "Hozidenta")

anana.descripcion_restaurante()

anana.abrir_restaurante()

nana = Restaurante("Rick", "Classico")

nana.descripcion_restaurante()

nana.abrir_restaurante()
```
