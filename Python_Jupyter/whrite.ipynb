```python
from pathlib import Path

def bananas_locas():
    
    datos = []
    while True:
        
        banana = input("Tu nombre es: \n")

        if banana.lower() == "q":
            break
        
        datos.append(banana)
        
    return datos




Listado = bananas_locas()

contenido = "\n".join(Listado)

path = Path("p.txt")
path.write_text(contenido)

print("--- Proceso finalizado. Datos guardados. ---")
```

```python
from pathlib import Path

def recopilar_nombres():
    datos = []  # La lista debe estar FUERA del bucle para que guarde todo
    
    while True:
        entrada = input("Introduce un nombre (o 'q' para salir): \n")
        
        if entrada.lower() == 'q':
            break
        
        datos.append(entrada) # Guardamos cada entrada en la lista
    
    return datos

# 1. Llamamos a la función y guardamos el resultado
lista_final = recopilar_nombres()

# 2. Convertimos la lista a una sola cadena de texto (String)
# Usamos "\n".join() para que cada nombre esté en una línea distinta
contenido = "\n".join(lista_final)

# 3. Escribimos al archivo
path = Path("pi_digits.txt")
path.write_text(contenido)

print("--- Proceso finalizado. Datos guardados. ---")
```
