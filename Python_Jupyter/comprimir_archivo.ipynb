```python
import zipfile
import os

# CONFIGURACIÓN DEL INGENIERO KEVIN
# Usamos 'r' al principio para que Windows no se confunda con las barras \
archivo_a_mover = r"C:\Users\Usuario\Downloads\fotos, etc1.zip" 
nombre_destino = "backup_final_60gb.zip"

def comprimir_archivo_grande(ruta_origen, nombre_zip):
    # Verificamos si el archivo realmente existe antes de empezar
    if not os.path.exists(ruta_origen):
        print(f"ERROR: No encontré nada en {ruta_origen}")
        print("Asegurate de que el nombre sea exacto.")
        return

    print(f"Iniciando compresión de: {ruta_origen}...")
    
    with zipfile.ZipFile(nombre_zip, 'w', zipfile.ZIP_DEFLATED, allowZip64=True) as zipf:
        zipf.write(ruta_origen, os.path.basename(ruta_origen))
        
    print(f"¡Éxito! Archivo guardado como: {nombre_zip}")

comprimir_archivo_grande(archivo_a_mover, nombre_destino)
```

```python
import os

def unir_archivo(nombre_base_zip, cantidad_partes):
    with open(nombre_base_zip, 'wb') as archivo_final:
        for i in range(1, cantidad_partes + 1):
            nombre_parte = f"{nombre_base_zip}.part{i}"
            print(f"Uniendo {nombre_parte}...")
            with open(nombre_parte, 'rb') as parte:
                archivo_final.write(parte.read())
    print(f"¡Éxito! Tu archivo original '{nombre_base_zip}' ha sido reconstruido.")

# CONFIGURACIÓN
# Poné el nombre exacto del archivo que querés recuperar
unir_archivo("backup_final_60gb.zip", 1) # Cambiá el 4 por la cantidad
```
