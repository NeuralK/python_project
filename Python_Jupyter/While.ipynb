```python
ingredientes = "\n¿Que ingredientes le agregamos a su pizza?: "

ingredientes += "\nEscriba 'listo' para finalizar la lista de ingredientes"

while True:

    lista = input(ingredientes)

    if lista == "listo":
        print("Bien, se la preparamos...")
        break
    else:
        print(f"Anotado en su pizza le ponemos: {lista}")
```
