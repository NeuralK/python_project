```python
class Electric_Car:

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.battery = Battery()
        
    def describe_car(self):    
        Lists = [self.make, self.model, self.year]
        for list in Lists:
            print(list)

       
class Battery:
    def __init__(self, battery_size=40):
        # Asegurate de que el nombre sea el mismo en ambos lados
        self.battery_size = battery_size 

    def describe_battery(self):
        # Ahora sí, self.battery_size existe
        print(f"This car has a {self.battery_size}-kwh battery.")

    def upgrade_battery(self):
        if self.battery_size < 65:
            self.battery_size = 65
            print(f"This car has a {self.battery_size}-kwh battery.")
        else:
            print(f"This car has a {self.battery_size}-kwh battery.")




banana = Electric_Car("Chevrlolet", "Onix LS", 2026)
banana.describe_car()
banana.battery.describe_battery()
banana.battery.upgrade_battery()
```

```python
class Usuario:

    def __init__(self, nombre, apellido, age):
        
        self.nombre = nombre

        self.apellido = apellido

        self.age = int(age)

        self.intentos_inicios = 0

    def describir_usuario(self):

        print(f"\nAsí que usted se llama {self.nombre}.")
        print(f"\nMmmh te apellidas {self.apellido} raro apellido.") 
        print(f"\nAsí que tienes {self.age} es curioso.")
    

    def saludar(self):

        self.saludo = "\nBienvenido"

        print(self.saludo.title())

    def incrementar_intentos_inicios(self):

        self.intentos_inicios += 1
        print(f"Numero de intentos: {self.intentos_inicios}")

    def restablecer_intentos_inicio(self):
        
        self.intentos_inicios = 0
        print(f"Numero de intentos: {self.intentos_inicios}")


class Admin(Usuario):
    def __init__(self, nombre, email, age): # O los parámetros que pida Usuario
        super().__init__(nombre, email, age) # Esto conecta con la clase padre
            
               # Atributo específico de Admin

        self.privilegios = Privilegios()



class Privilegios:
    def __init__(self):
        self.privilegios = ["Puede eliminar usuarios", "Puede añadir comentarios", "Puede borrar comentarios"]
    def show_privileges(self):

        for pri in self.privilegios:
            print(pri)



            
```

```python
class Restaurante:

    def __init__(self, nombre_restaurante, tipo_cocina):
        self.nombre_restaurante = nombre_restaurante
        self.tipo_cocina = tipo_cocina
        self.numero_servido = 0

    def descripcion_restaurante(self):

        print(f"Así que el restaurante se llama {self.nombre_restaurante}.")
        print(f"\nAsí que {self.tipo_cocina} es el tipo de cocina de usan.")
    
    def abrir_restaurante(self):
        self.open = "\nRestaurante abierto"
        
        print(self.open.title())

    def restaurante(self):

        print(f"\nAsi que tienen {self.numero_servido} clientes servidos.")
    
    def establecer_numero_servido(self, actualizacion):

        self.numero_servido = actualizacion

    def incrementar_numero_servidos(self, incremento):

        self.numero_servido += incremento



banana = Restaurante("Buiton", "Oriental")

banana.descripcion_restaurante()

banana.abrir_restaurante()

banana.numero_servido = 21


banana.establecer_numero_servido(25)

banana.incrementar_numero_servidos(100)

banana.restaurante()

class CarritoDeHelados(Restaurante):

    def __init__(self, nombre_restaurante, tipo_cocina):
        super().__init__(nombre_restaurante, tipo_cocina)

        print(f"\nAsí que el restaurante se llama {self.nombre_restaurante}.")
        print(f"\nAsí que {self.tipo_cocina} es el tipo de cocina de usan.")

    def sabores(self):

        print("En los sabores de helado hay: \nBanana \nSandia \nLimon")

bananas = CarritoDeHelados("Grido", "artesanal")

bananas.sabores()
```
