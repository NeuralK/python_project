```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Conexión al Chrome abierto
chrome_options = Options()
chrome_options.add_experimental_option("debuggerAddress", "127.0.0.1:9222")
driver = webdriver.Chrome(options=chrome_options)

rut_objetivo = "217445090011"

try:
    # 1. Filtrar por el RUT
    # Reemplaza 'ID_DEL_BUSCADOR' por el que encuentres con F12
    buscador = driver.find_element(By.ID, "ID_DEL_BUSCADOR") 
    buscador.clear()
    buscador.send_keys(rut_objetivo)
    buscador.send_keys(Keys.RETURN)
    time.sleep(3) # Esperar a que cargue la lista filtrada

    # 2. Encontrar todos los botones de descarga
    # Aquí es donde ocurre la magia. 'By.CLASS_NAME' suele ser 
    # algo como 'btn-download' o un icono de PDF.
    botones_descarga = driver.find_elements(By.CLASS_NAME, "CLASE_DEL_BOTON_DESCARGA")

    print(f"Se encontraron {len(botones_descarga)} boletas. Iniciando descarga...")

    for boton in botones_descarga:
        boton.click()
        time.sleep(1) # Pausa para no saturar las descargas
        
    print("¡Proceso terminado!")

except Exception as e:
    print(f"Error: {e}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Conexión al Chrome que ya tenés abierto
chrome_options = Options()
chrome_options.add_experimental_option("debuggerAddress", "127.0.0.1:9222")
driver = webdriver.Chrome(options=chrome_options)

rut_objetivo = "217445090011"

try:
    print(f"Buscando y resaltando boletas para: {rut_objetivo}...")
    
    # Buscamos todos los elementos que contienen el RUT
    elementos = driver.find_elements(By.XPATH, f"//*[contains(text(), '{rut_objetivo}')]")
    
    resaltados = 0
    for el in elementos:
        try:
            # Subimos hasta encontrar el contenedor de la fila (generalmente un 'tr' o un 'div' grande)
            # Aplicamos el resaltado directamente con JavaScript
            driver.execute_script(
                "arguments[0].style.backgroundColor = '#FFFF00';" # Amarillo
                "arguments[0].style.border = '3px solid red';"     # Borde rojo
                "arguments[0].style.padding = '5px';", 
                el
            )
            resaltados += 1
        except:
            continue
            
    print(f"¡Listo! Se resaltaron {resaltados} boletas en tu pantalla.")
    print("Revisá la pestaña de Memory, deberían estar todas marcadas.")

except Exception as e:
    print(f"Error al intentar resaltar: {e}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Conexión al Chrome abierto
chrome_options = Options()
chrome_options.add_experimental_option("debuggerAddress", "127.0.0.1:9222")
driver = webdriver.Chrome(options=chrome_options)

rut_objetivo = "217445090011"

def buscar_rut_en_toda_la_pagina():
    print(f"Iniciando escaneo profundo para el RUT: {rut_objetivo}...")
    encontrado = False
    
    # 1. Buscamos en el texto plano de TODA la página (lo más básico)
    if rut_objetivo in driver.page_source:
        print("¡ÉXITO! El RUT existe en el código fuente de la página.")
        encontrado = True
    else:
        print("El RUT no aparece ni siquiera en el código fuente principal.")

    # 2. Buscamos específicamente en elementos de tabla por si el page_source falló
    celdas = driver.find_elements(By.XPATH, f"//*[contains(text(), '{rut_objetivo}')]")
    print(f"Se encontraron {len(celdas)} elementos físicos que contienen el RUT.")
    
    if len(celdas) > 0:
        for i, celda in enumerate(celdas):
            print(f"Elemento {i+1} detectado en el tag: <{celda.tag_name}>")

buscar_rut_en_toda_la_pagina()
```
