```python
def usuario(nombre, apellido, *todo):

    nombre_completo = nombre, apellido
    for name in nombre_completo:
        if name == nombre:
            print(f"Interesante... {name}")
        else:
            print(f"Curioso..... {name}")
    for all in todo:
        print(f"Interesante {all}, cuenteme más.")

usuario("Kevin", "Moreno", "zapato", "bondiola")
```

```python
def saludar(names, guardados):

    while names:
        proceso = names.pop()
        print(f"Hola, {proceso}")
        guardados.append(proceso)

def mostrar(guardados):

    for guardado in guardados:
        print(guardado)

names = ["salame", "banana", "inutil"]

guardados = []

saludar(names[:], guardados)

mostrar(guardados)

print(guardados)
print(names)
```
