```python
def arithmetic_arranger(problems, show_answers=False):
    
    pepe = []
    zape = []
    jojo = []
    magi = []
    
    if len(problems)  > 5:
                return "Error: Too many problems."
    
    for problem in problems:
    
    
        
        if "*" in problem or "/" in problem:
                return "Error: Operator must be '+' or '-'."
        
        count = problem.split()
        n1, op, n2 = count[0], count[1], count[2]
        
        if len(count[0]) > 4 or len(count[2]) > 4:
                return 'Error: Numbers cannot be more than four digits.'
        
        if not count[0].isdigit() or not count[2].isdigit():
                return 'Error: Numbers must only contain digits.'
        
        wirly = max(len(n1), len(n2)) + 2
        pepe.append(n1.rjust(wirly))
        zape.append(op + n2.rjust(wirly - 1))
        jojo.append("-" * wirly)

        if show_answers:
            res = str(int(n1) + int(n2) if op == "+" else int(n1) - int(n2))
            magi.append(res.rjust(wirly))
            
    fila1 = "    ".join(pepe) 
    fila2 = "    ".join(zape)
    fila3 = "    ".join(jojo)  

    resultado_final = fila1 + "\n" + fila2 + "\n" + fila3

    if show_answers:
        fila4 = "    ".join(magi)
        resultado_final += "\n" + fila4
        
    return resultado_final
            
               

    
print(f'\n{arithmetic_arranger(["32 + 698", "3801 - 2", "45 + 43", "123 + 49"])}')

    
```
