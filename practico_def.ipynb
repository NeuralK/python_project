```python
import os
from PyPDF2 import PdfMerger

# 1. Esto detecta tu usuario de Windows automáticamente
usuario_actual = os.getlogin() 
ruta_real = os.path.join(r'C:\Users', usuario_actual, 'Desktop', 'ASDAS')

nombre_salida = 'PDF_Unificado_Final.pdf'

def unir_pdfs(directorio, salida):
    # Verificamos la ruta final generada
    if not os.path.exists(directorio):
        print(f"ERROR: No encuentro la carpeta en: {directorio}")
        print("Fijate si la carpeta ASDAS está realmente en el escritorio.")
        return

    merger = PdfMerger()
    archivos = [f for f in os.listdir(directorio) if f.endswith('.pdf')]
    
    # Ordenamos para que (1), (10), (11) no se mezclen (orden natural)
    archivos.sort(key=lambda f: int(''.join(filter(str.isdigit, f)) or 0))

    print(f"--- Iniciando unión de {len(archivos)} archivos ---")

    for archivo in archivos:
        ruta_completa = os.path.join(directorio, archivo)
        merger.append(ruta_completa)

    # El archivo final se guarda en la carpeta donde está tu Jupyter
    merger.write(salida)
    merger.close()
    print(f"\n--- ÉXITO: Archivo guardado como {salida} ---")

unir_pdfs(ruta_real, nombre_salida)
```
