```python
def analizar_accion(nombre_accion, precio_apertura, precio_cierre):
    ganancia = precio_cierre - precio_apertura
    porcentaje = (ganancia / precio_apertura) * 100
    print(f"--- Reporte de {nombre_accion} ---")
    print(f"Ganancia neta: ${ganancia:.2f}")
    print(f"Rendimiento: {porcentaje:.2f}%")
    if ganancia > 0:
     print("resultado: ¡Día de Profit!")
    else:
      print("resultado: Pérdida.") 
analizar_accion("Apple", 150.2, 155.8)
print("-" * 30) 
analizar_accion("Tesla", 240.00, 235.50)
```

```python
acciones = [
    {"nombre": "Apple", "abre": 150.2, "cierra": 155.8},
    {"nombre": "Google", "abre": 140.0, "cierra": 148.2},
    {"nombre": "Tesla", "abre": 240.0, "cierra": 235.5},
    {"nombre": "Amazon", "abre": 180.0, "cierra": 178.0}
]

for a in acciones:
    analizar_accion(a["nombre"], a["abre"], a["cierra"])
```

```python
# --- CHULETA TÉCNICA DE LISTAS PARA EL ADMINISTRADOR ---

# 1. CREACIÓN (El estante de datos)
clientes = ["Ana", "Roberto", "Lucía", "Spam_Bot", "Marcos"]

# 2. MEDIR (Reporte de cantidad)
total_clientes = len(clientes)
print(f"Tienes {total_clientes} clientes en la lista.")

# 3. ACCESO (Indexación)
# Recordar: primer lugar es 0
primero = clientes[0]
ultimo = clientes[-1]
print(f"El primero es {primero} y el último es {ultimo}")

# 4. REBANAR (Slicing)
# Sacar los 2 primeros para prioridad
prioridad = clientes[:2]
# Sacar los 2 últimos que se unieron
recientes = clientes[-2:]
print(f"Prioridad: {prioridad} | Recientes: {recientes}")

# 5. MODIFICAR (Gestión de base de datos)
clientes.append("Nuevo_Cliente_VIP") # Agregamos al final
clientes.remove("Spam_Bot")        # Limpiamos el error
print(f"Lista actualizada: {clientes}")

# 6. MATEMÁTICA RÁPIDA (Para facturación)
pagos = [150.50, 200.00, 50.25]
total_dinero = sum(pagos)
print(f"Total a cobrar hoy: ${total_dinero}")
```
